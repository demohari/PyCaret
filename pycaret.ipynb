```python
print("No. of columns containing null values")
print(len(data1.columns[data1.isna().any()]))

print("No. of columns not containing null values")
print(len(data1.columns[data1.notna().all()]))

print("Total no. of columns in the data1frame")
print(len(data1.columns))
```

```python
#For getting Columns having at least 1 null value. (column names)
print(f"columns with null values : {list(data1.columns[data1.isnull().any()])}")

#For getting Columns with count, with having at least 1 null value.
print(f"columns with null values count : \n{data1[data1.columns[data1.isnull().any()]].isnull().sum()}")

#For getting percentage of the null count.
print(f"columns with null values percentage : \n{data1[data1.columns[data1.isnull().any()]].isnull().sum() * 100 / data1.shape[0]}")
```
